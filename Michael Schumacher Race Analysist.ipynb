```python
import chart_studio.plotly as py
import plotly.graph_objs as go
import plotly.figure_factory as FF
from matplotlib import pyplot as plt
from matplotlib import figure 
from datetime import datetime

import numpy as np
import pandas as pd

driverId=30

driverStandings = pd.read_csv(r"E:\TUGAS_KULIAH\probstok\project akhir\F1\driverStandings.csv")

standings=driverStandings.loc[driverStandings['driverId']==driverId]
print('Total lifetime race= ', standings.shape[0],' races')
y=standings['position'].values
count1st = np.count_nonzero(y == 1)
count2nd = np.count_nonzero(y == 2)
count3rd = np.count_nonzero(y == 3)
countOther = np.count_nonzero(y > 3)
labels = '1st place','2nd place','3rd place','Other'
arrayCount = [count1st, count2nd, count3rd, countOther]
explode = (0.1, 0, 0, 0)
fig1, ax1 = plt.subplots()
ax1.pie(arrayCount, explode=explode, labels=labels, autopct='%1.1f%%',
       shadow=True, startangle=90)
ax1.axis('equal')

plt.title('Michael Schumacher podium chart')
plt.show()
```

<p>As you can see Michael Schumacher not always come in 1st, place, but most of his race are placed in 1st place, even more than 50% of them are in podium (podium is a term for position 1st, 2nd, and 3rd).</p><p>But how many times Michael Schumacher not complete his race?</p>

```python
import chart_studio.plotly as py
import plotly.graph_objs as go
import plotly.figure_factory as FF
from matplotlib import pyplot as plt
from matplotlib import figure 
from datetime import datetime

import numpy as np
import pandas as pd


results = pd.read_csv(r"E:\TUGAS_KULIAH\probstok\project akhir\F1\results.csv")
results['position'] = results['position'].fillna(0)

results=results.loc[(results['driverId'] == 30) & (results['statusId'] > 1) & (results['position'] == 0)]
print('Total DNF :', results.shape[0],' races')
y=results['statusId'].values
countDisqualified = np.count_nonzero(np.logical_or(y==2, y==81))
countCollision1 = np.count_nonzero(y == 3)  
countCollision2 = np.count_nonzero(y == 4) 
countCollision3 = np.count_nonzero(y == 104)
countCollision = countCollision1 + countCollision2 + countCollision3
countEngine1 = np.count_nonzero(np.logical_and(y> 11, y < 14)) 
countEngine2 = np.count_nonzero(np.logical_and(y> 19, y < 45)) 
countEngine3 = np.count_nonzero(np.logical_and(y> 45, y < 50))
countEngine4 = np.count_nonzero(y == 51) 
countEngine5 = np.count_nonzero(y == 56) 
countEngine6 = np.count_nonzero(y == 129) 
countEngine7 = np.count_nonzero(np.logical_and(y> 58, y < 77)) 
countEngine8 = np.count_nonzero(np.logical_and(y> 77, y < 88)) 
countEngine9 = np.count_nonzero(np.logical_and(y> 92, y < 111))
countEngine = countEngine1 + countEngine2 + countEngine3 + countEngine4 + countEngine5 +countEngine6 + countEngine7 + countEngine8 + countEngine9
count = countDisqualified + countCollision + countEngine
countOther = results.shape[0] - count
labels = 'Disqualified', 'Collision', 'Engine Problem', 'Misc.'
arrayCount = [countDisqualified, countCollision, countEngine, countOther]
explode = (0, 0, 0, 0)
fig1, ax1 = plt.subplots()
ax1.pie(arrayCount, explode=explode, labels=labels, autopct='%1.1f%%',
       shadow=True, startangle=90)
ax1.axis('equal')
```

<p>As we can see Michael Schumacher mainly DNF because of engine problem. Is it his constructor was so bad?</p>
